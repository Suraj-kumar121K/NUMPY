```python
import numpy as np

a = np.array([1, 2, 3, 4, 5, 6])
b = np.array([7, 8, 9, 10, 11, 12])
```

# ➡️ Manipulation in NumPy
- Modifying the structure, order, shape, or elements of a NumPy array.

## Operations Included in NumPy Manipulation

### 1️⃣ Reshape / Resize
- Change the shape or size of an array.


```python
a.reshape(3, 2)
```

## 2️⃣ Flattening
- Convert a multi-dimensional array into a 1D array.

### returns a copy      

```python
a.flatten() 
```

### returns a view (memory efficient)

```python
a.ravel()
```

## 3️⃣ Joining Arrays
- Combine multiple arrays into one

```python
np.concatenate((a, b), axis=0)
```

```python
np.hstack((a, b))
```

```python
np.vstack((a, b))
```

## 4️⃣ Splitting Arrays
- Divide an array into multiple sub-arrays.

```python
np.split(a, 3)
```

# 5️⃣Adding / Removing Elements

```python
np.append(a, 10)
```

```python
np.insert(a, 1, 99)
```

```python
np.delete(a, 2)
```

# 6️⃣ Changing Data Type

```python
a.astype(float)
```

# 7️⃣ Copy vs View
- Memory management in NumPy.

# creates a new array

```python
a.copy()
```

# shares the same memory

```python
a.view()
```
