# Broadcasting in NumPy
- Broadcasting is a NumPy feature that allows element-wise operations on arrays of different     shapes without writing explicit loops.

# ðŸ”¹ Why Broadcasting?
- Removes the need for loops
- Faster computation
- Clean and readable code
- Memory efficient

## ðŸ”¹ Broadcasting Rules (IMPORTANT)
- NumPy compares array shapes from right to left.
- Two dimensions are compatible if:
1. They are equal, or
2. One of them is 1

## Broadcasting Example

```python
import numpy as np

a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

a + b
```
