# What is Random in NumPy
- The `numpy.random` module is used to generate random numbers or arrays.

- Can create random floats, integers, normal distributions, or random selections.

### Examples:

1. Random floats [0,1)
- Generates random floating-point numbers in the range 0 (inclusive) to 1 (exclusive).

- NumPy, use `np.random.rand()`.

```python
import numpy as np
arr = np.random.rand(5)
print(arr)
```

2. Random integers
- Generates random integer numbers within a specified range.

- use np.random.randint(low, high, size).

### Syntax
`np.random.randint(low, high, size)`
- low → minimum integer (inclusive)
- high → maximum integer (exclusive)
- size → number of integers to generate

```python
arr = np.random.randint(1, 10, size=5)  # 5 random integers between 1 and 9
print(arr)
```

```python
arr = np.random.randint(1, 10, size=5)
print(arr)
```

3. Random from normal distribution
- Generates random numbers that follow a normal (Gaussian) distribution.

- Normal distribution = bell-shaped curve with mean = 0 and standard deviation = 1 by default.

```python
arr = np.random.randn(5)
print(arr)
```

4. Random choice from a list
- `np.random.choice()` selects one or more random elements from a given list or array.

```python
arr = np.random.choice([10, 20, 30, 40], size=3)
print(arr)
```
