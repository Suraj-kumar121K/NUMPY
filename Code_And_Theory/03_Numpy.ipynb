# Numpy
- import Numpy

```python
import numpy as np
```

1. From existing data (most common)

```python
# 1D array (vector)
arr1 = np.array([1, 2, 3, 4])          # → array([1, 2, 3, 4])

# 2D array (matrix)
arr2 = np.array([[1, 2, 3],
                [4, 5, 6]])           # shape → (2, 3)

# From tuple / nested list
arr3 = np.array((10, 20, 30))          # same as list
```

- Create List And Array

```python
List = [10, 'S', 20, 30, 40]
arr = np.array(List)
print(arr)
type(arr)
```
