## Import data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"C:\Users\Anish Kumar\Downloads\students.xlsx")
df
```

### NumPy Array Banaao

```python
marks = df[["Math", "Science", "English"]].to_numpy()
marks
```

### Total Calculate

```python
df["Total"] = marks.sum(axis=1)
```

### Average Calculate

```python
df["Average"] = marks.mean(axis=1)
```

### Pass / Fail

```python
df["Result"] = np.where((marks < 33).any(axis=1), "Fail", "Pass")
```

## Grade

```python
df["Grade"] = np.where(df["Average"] >= 90, "A+",
               np.where(df["Average"] >= 75, "A",
               np.where(df["Average"] >= 60, "B",
               np.where(df["Average"] >= 40, "C", "D"))))
```

## Rank

```python
df["Rank"] = df["Total"].rank(ascending=False).astype(int)
```
