# NumPy Array Attributes

1️⃣ ndim(Number of Dimensions)
shows the dimensionality of a NumPy array.
- 1 → 1D array
- 2 → 2D array
- 3 → 3D array

```python
import numpy as np

# create a 2-D array 
array1 = np.array([[2, 4, 6],
                  [1, 3, 5]])

# check the dimension of array1
print(array1.ndim) 
```

2️⃣ size :- gives the total count of elements in the array

```python
import numpy as np

array1 = np.array([[1, 2, 3],
                 [6, 7, 8]])

# return total number of elements in array1
print(array1.size)
```

3️⃣ dtype (Data Type)
- dtype shows the data type of elements stored in the array.
- NumPy arrays are homogeneous (same data type).

```python
import numpy as np

# create an array of integers 
array1 = np.array([6, 7, 8])

# check the data type of array1
print(array1.dtype) 
```

4️⃣ shape (Array Structure)
- shape tells the structure of the array in (rows, columns) form

```python
import numpy as np

array1 = np.array([[1, 2, 3],
                [6, 7, 8]])

# return a tuple that gives size of array in each dimension
print(array1.shape)
```

5️⃣ itemsize (Memory per Element)
- itemsize tells how many bytes one element occupies in memory.

Examples:
- int32 → 4 bytes
- int64 → 8 bytes

```python
import numpy as np

# create a default 1-D array of integers
array1 = np.array([6, 7, 8, 10, 13])

# create a 1-D array of 32-bit integers 
array2 = np.array([6, 7, 8, 10, 13], dtype=np.int32)

# use of itemsize to determine size of each array element of array1 and array2
print(array1.itemsize)  
print(array2.itemsize)  
```

6️⃣ data (Memory Location)


- data gives the memory buffer address where array data is stored.

- Used mainly for low-level memory access, not common in daily coding.

```python
import numpy as np
array1 = np.array([6, 7, 8])
array2 = np.array([[1, 2, 3],
                   	    [6, 7, 8]])

# print memory address of array1's and array2's data
print("\nData of array1 is: ",array1.data)
print("Data of array2 is: ",array2.data)
```
